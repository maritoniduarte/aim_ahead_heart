```python
(other_mhdata2018
 .drop(columns=['YEAR', 'AGE', 'EDUC', 'ETHNIC', 'RACE', 'GENDER', 'OPISERVICE', 'RTCSERVICE', 'IJSSERVICE', 'MH2', 'MH3', 'SUB', 'MARSTAT', 'SMISED', 'SAP', 'EMPLOY', 'DETNLF', 'VETERAN', 'LIVARAG', 'NUMMHS', 'TRAUSTREFLG', 'ANXIETYFLG', 'ADHDFLG', 'CONDUCTFLG', 'DELIRDEMFLG', 'BIPOLARFLG', 'DEPRESSFLG', 'ODDFLG', 'PDDFLG', 'PERSONFLG', 'SCHIZOFLG', 'ALCSUBFLG', 'OTHERDISFLG', 'DIVISION', 'REGION', 'CASEID'])
 .groupby('MH1').sum()
 .plot.bar()
)
```

```python
other_sphservice2018 = other_mhdata2018.groupby('MH1')['SPHSERVICE'].count()
result_other_sphservice2018 = other_mhdata2018.groupby(by=['MH1','SPHSERVICE']).size().reset_index()
result_other_sphservice2018.rename(columns={0:"count"}, inplace=True)

# plot bar graph
result_other_sphservice2018.set_index(["MH1","SPHSERVICE"])['count'].unstack().plot.bar()
```

```python
other_cmpservice2018 = other_mhdata2018.groupby('MH1')['CMPSERVICE'].count()
result_other_cmpservice2018 = other_mhdata2018.groupby(by=['MH1','CMPSERVICE']).size().reset_index()
result_other_cmpservice2018.rename(columns={0:"count"}, inplace=True)

# plot bar graph
result_other_cmpservice2018.set_index(["MH1","CMPSERVICE"])['count'].unstack().plot.bar()
```
